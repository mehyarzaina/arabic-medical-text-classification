# **Preprocessing**

```python
from deep_translator import GoogleTranslator
import re

def contains_english(text):
    return any(re.search(r'[A-Za-z]', word) for word in text.split())

def translate(text):
    if contains_english(text):
        try:
            return GoogleTranslator(source='auto', target='ar').translate(text)
        except Exception as e:
            print("Translation error:", e)
            return text
    return text


df['question_body'] = df['question_body'].apply(translate)
```

```python
import re
from farasa.stemmer import FarasaStemmer

stemmer = FarasaStemmer(interactive=True)


def preprocess_text(text):

    text = re.sub(r'[إأآ]', 'ا', text)
    text = re.sub(r'ؤ', 'و', text)
    text = re.sub(r'ئ', 'ي', text)
    text = re.sub(r'ء', '', text)
    text = re.sub(r'ة', 'ه', text)

    # Remove harakat
    text = re.sub(r'[\u064B-\u0652]', '', text)

    text = re.sub(r'[،.:؛؟!*,&)@-_=+#(-/\|)]', '', text)

    #remove numbers
    text = re.sub(r'\d+', '', text)

    # Remove English letters
    text = re.sub(r'[A-Za-z]', '', text)

    words = word_tokenize(text)

    my_stop_words = ['السلام', 'عليكم', 'Tawuniya']
    arabic_stopwords = set(stopwords.words('arabic') + my_stop_words)
    words = [word for word in words if word not in arabic_stopwords]

    words = [stemmer.stem(word) for word in words]

    return words
```

```python
cleaned_text = [preprocess_text(proverb) for proverb in df['question_body'] ]
df['question_body_1'] = cleaned_text
```
